```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(-1, 2, 100)
y = np.linspace(-1, 2, 100)

x1, y1 = np.meshgrid(x, y)
fig, ax = plt.subplots()
z = np.sin(np.cos(np.tan(x1))) * np.sin(np.cos(np.tan(y1)))

ax.contourf(x1, y1, z, colors=['yellow', 'red', 'blue'])


```

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 2*np.pi, 500)

fig, axes = plt.subplots(1, 2)

x = np.sin(4*t) - np.cos(t/2)
y = np.sin(t/2) - np.cos(4*t)

axes[0].fill(x, y,
                color = 'darkgreen',
                fill = False)
axes[0].patch.set_facecolor('floralwhite')
axes[0].set_title('?', 
             fontweight ="bold",
                 fontsize = 18)

axes[1].fill(x, y,
                color = 'darkgreen')
axes[1].set_title('!', 
             fontweight ="bold", 
                  fontsize = 18)
axes[1].patch.set_facecolor('floralwhite')

fig.set_facecolor("limegreen")

fig.set_figwidth(12)
fig.set_figheight(7)    

plt.show()
```
