## Exercise 1  
1. Создайте график $x^2⋅sin(1/x^2)+x$ на интервале $[−1,1]$, используя 250 точек. Не забудьте пометить оси!
2. Создайте semilogy график (с логарифмической шкалой для оси y и линейной шкалой для оси x) относительной разницы  функций $1/(1+x^2)$ и $1/x^2$ на интервале $[5,25]$. (Относительная разница чисел $a$ и $b$ определяется как $|1−a/b|$)
3. В Scipy также есть много менее распространенных специальных функций. Взгляните на обширный список <a href=https://docs.scipy.org/doc/scipy/reference/special.html> здесь</a>! Выберите несколько интересных спец функций, которые вы знаете (или, может быть, не знаете), подставьте несколько чисел и начертите их.


```python
#1
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

x = np.linspace(-1,1,250)
y = x**2 * np.sin(1/(x**2)) + x
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('функция sin')
```

```python
#2

x = np.linspace(5, 25, 250)
y1 = 1 / (1 + x**2)
y2 = 1 / x**2
y = 1 - y1 / y2

fig, ax = plt.subplots(1,2)

ax[0].plot(x, y)
ax[0].set_xlabel('x')
ax[0].set_ylabel('y(x)')
ax[1].semilogy(x, y)
ax[1].set_xlabel('x')
ax[1].set_ylabel('y(x)')
```

```python
#3
import scipy.special

x = np.linspace(-2,4,100)

y1 = scipy.special.kolmogorov(x)
y2 = scipy.special.zetac(x)

fig, ax = plt.subplots(1,2)

ax[0].plot(x, y1, color = 'black', ls = '-', label='Kolmogorov')
ax[0].set_xlabel('y')
ax[0].set_ylabel('x')
ax[0].legend()
ax[1].plot(x, y2, color = 'orange', ls = '-.', label='Zetac')
ax[1].set_xlabel('x')
ax[1].set_ylabel('y')
ax[1].legend()
```

## Exercise 2  
1. Построить график функции $sin(x)/x$ на отрезке $[-6\pi;6\pi]$ и добавить тики по оси $x$ с шагом $\pi/2$ от $-6\pi$ до $6\pi$, причём обозначения тиков должны содержать греческую букву $\pi$. Добавить сетку на график.

```python
x = np.linspace(-6 * np.pi, 6 * np.pi, 100)
y = np.sin(x) / x

fig, ax = plt.subplots(figsize=(15,4))

ax.plot(x, y)
plt.xticks(np.linspace(-6*np.pi, 6*np.pi, 25), [r'$-6\pi$', r'$-11\pi/2$', r'$-5\pi$', r'$-9\pi/2$', r'$-4\pi$', r'$-7\pi/2$', 
                                               r'$-3\pi$', r'$-5\pi/2$', r'$-2\pi$', r'$-3\pi/2$', r'$-\pi$', r'$-\pi/2$',
                                               r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$', r'$5\pi/2$', r'$3\pi$', 
                                               r'$7\pi/2$', r'$4\pi$', r'$9\pi/2$', r'$5\pi$', r'$11\pi/2$', r'$6\pi$'])
    
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid()
```

## Exercise 3  
1. Создайте цветовой график $sin(x)sin(y)$ на интервале $[−π,π]×[−π,π]$.
2. Создайте цветовой график $e^{-x^2-y^2}$ на интервале $[−3,3]×[−3,3]$.

```python
#1
x = np.linspace(-np.pi, np.pi, 100)
y = np.linspace(-np.pi, np.pi, 100)
x1, y1 = np.meshgrid(x, y)
fig, ax = plt.subplots(figsize=(5,5))
ax.contourf(x1 , y1, np.sin(x1) * np.sin(y1), cmap = 'plasma', alpha = 0.7)
```

```python
#2
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
x1, y1 = np.meshgrid(x, y)
fig, ax = plt.subplots()
z = np.exp(- y1**2 - x1**2)
ax.contourf(x1 , y1, z, colors=['red', 'white', 'blue'], extend='both')
```
